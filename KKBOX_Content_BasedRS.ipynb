```python
def nearestneighbours(songid, K):
    allsongs = pd.DataFrame(songs['song_id'])
    allsongs = allsongs[allsongs['song_id']!=songid]
    print(allsongs)
    songs_reindex = songs.set_index('song_id')
    allsongs['distance'] = allsongs['song_id'].apply(lambda x: hamming(songs_reindex.loc[songid], songs_reindex.loc[x]))
    Knearestsongs = allsongs.sort_values(['distance'],ascending= True)['song_id'][:K]
    return Knearestsongs
```

```python
def nearestneighbours(userid, songid, K=15):
    print("Song: ", songs_info_all[songs_info_all['song_id']==songid]['name'], "Artist: ", songs_info_all[songs_info_all['song_id']==songid]['artist_name'])
    
    allsongs = pd.DataFrame(songs['song_id'])
    allsongs = allsongs[allsongs['song_id']!=songid]
#     print(allsongs)
    #calucate the nearest songs based on the one hot encoded features, return it into the allsongs df
    songs_reindex = songs.set_index('song_id')
    allsongs['distance'] = allsongs['song_id'].apply(lambda x: hamming(songs_reindex.loc[songid], songs_reindex.loc[x]))
    Knearestsongs = allsongs.sort_values(['distance'],ascending= True)['song_id']
#     print(Knearestsongs[:K])
    #keep the songs that is not played by the user before
    Knearestsongs = pd.DataFrame(Knearestsongs)
    Knearestsongs.columns= ['song_id']
    Knearestsongs = Knearestsongs[~Knearestsongs['song_id'].isin(users_record[users_record['msno']==userid]['song_id'])]
    Knearestsongs = Knearestsongs[:K]
    print(f"result: {Knearestsongs}")
    return Knearestsongs
    
```

# Return N recommendated songs based on a list of songs (for slides)

```python
def nearestneighbours_list_of_songs(userid, list_of_songs, K=20):
    list_of_recommendations = []
    for songid in list_of_songs:
        print("Song: ", songs_info_all[songs_info_all['song_id']==songid]['name'], "Artist: ", songs_info_all[songs_info_all['song_id']==songid]['artist_name'])
    #create a dataframe that stores songs id
        allsongs = pd.DataFrame(songs['song_id'])
        allsongs = allsongs[allsongs['song_id']!=songid]

    #calucate the nearest songs based on the one hot encoded features, return it into the allsongs df
        songs_reindex = songs.set_index('song_id')
        allsongs['distance'] = allsongs['song_id'].apply(lambda x: hamming(songs_reindex.loc[songid], songs_reindex.loc[x]))
        Knearestsongs = allsongs.sort_values(['distance'],ascending= True)['song_id']

    #keep the nearest songs that is not played by the user before
    #append the results generated by different songs in the same list
        Knearestsongs = pd.DataFrame(Knearestsongs)
        Knearestsongs.columns= ['song_id']
        Knearestsongs = Knearestsongs[~Knearestsongs['song_id'].isin(users_record[users_record['msno']==userid]['song_id'])]
        Knearestsongs = Knearestsongs[:K]
        print(f"result: {Knearestsongs}")
        for song in Knearestsongs['song_id']:
            list_of_recommendations.append(song)
        
    
    #return the list of songs which has showed up the most
        song_counter = {}
        for i in list_of_recommendations:
            if i in song_counter:
                song_counter[i] += 1
            else:
                song_counter[i] = 1

        top_songs = sorted(song_counter, key = song_counter.get, reverse = True)

        Knearestsongs_top_list = top_songs[:K]
        
        
    #print out the song names and artist names
        for i in Knearestsongs_top_list:
            song = songs_info_all[songs_info_all['song_id']==i]['name']
            artist = songs_info_all[songs_info_all['song_id']==i]['artist_name']
            print(f"Song: {song},Artist: {artist}")

        return list_of_recommendations
    
```
